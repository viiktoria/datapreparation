### import dataset and update columnnames

```python
import pandas as pd
data = pd.read_csv('../Jan 2021 Data - Viz5 Gender Inequality and HIVAIDS.csv', encoding = 'unicode_escape', engine ='python')
#ecoding='unicode_escape' has to be used to enable opening
data.info()
```

```python
data.columns = ['Country', 'Region', 'Year', 'Sex', 'Age',
       'Estimated rate new HIV infection per 1000 people',
       'Estimated annual AIDS related deaths',
       'Estimated annual new HIV infections',
       'Estimated number of people living with HIV',
       'Estimated rate annual AIDS related deaths per 100000 people']
```

```python
data.head(4)
```

### calculate mean values per year for female and male
outcoming dataframe: "mean_per_year_and_sex"

```python
mean_per_year_and_sex = data.groupby(by=['Year', 'Sex'], as_index=False).mean().round(3)
mean_per_year_and_sex[['Country', 'Region', 'Age']] = ['Mean_per_year_and_sex', 'None', 'Age 10-19']
mean_per_year_and_sex = mean_per_year_and_sex.sort_values(by=['Year','Sex'])
mean_per_year_and_sex.head(3)
```

### calculate mean for each year
outcoming dataframe:"mean_per_year"

```python
mean_per_year = data.groupby(by='Year', as_index=False).mean().round(3)

mean_per_year[['Country', 'Region', 'Age', 'Sex']] = ['Mean_per_year', 'None', 'Age 10-19', 'Mean']
mean_per_year.head(4)
```

### calculate the differnce between female and total average and male and total average

#### diff female- average
outcoming dataframe: "diff_female_mean"

```python
#concatinate df mean_per_year and mean_per_year_and_sex with 'female' to check, if the output is the expected:

temp = pd.concat([mean_per_year, mean_per_year_and_sex.loc[mean_per_year_and_sex['Sex']== 'Female']])\
.sort_values(by=['Year', 'Country'])
temp.head(4)
```

```python
#concatinate df mean_per_year and the 'female' rows of mean_per_year_and_sex 
# and group by year to calculate difference between mean and female numbers

diff_female_mean = pd.concat([mean_per_year, mean_per_year_and_sex.loc[mean_per_year_and_sex['Sex']== 'Female']])\
.sort_values(by=['Year','Country'])[['Year','Estimated rate new HIV infection per 1000 people','Estimated annual AIDS related deaths','Estimated annual new HIV infections','Estimated number of people living with HIV','Estimated rate annual AIDS related deaths per 100000 people']]\
.groupby('Year')\
.diff().dropna()
diff_female_mean[['Country', 'Region', 'Age', 'Sex']] = ['Diff_female_mean', 'None', 'Age 10-19', 'Diff_female_mean']

diff_female_mean = diff_female_mean.reset_index(drop=True)
diff_female_mean['Year'] = mean_per_year['Year']
diff_female_mean.head(4)
```

#### diff male-mean

```python
#concatinate df mean_per_year and mean_per_year_and_sex with 'male' to check, if the output is the expected:
pd.concat([mean_per_year, mean_per_year_and_sex.loc[mean_per_year_and_sex['Sex']== 'Male']])\
.sort_values(by=['Year', 'Country']).head(6)
```

```python
#concatinate df mean_per_year and the 'female' rows of mean_per_year_and_sex 
# and group by year to calculate difference between mean and male numbers

diff_male_mean = pd.concat([mean_per_year, mean_per_year_and_sex.loc[mean_per_year_and_sex['Sex']== 'Male']])\
.sort_values(by=['Year','Country'])[['Year','Estimated rate new HIV infection per 1000 people','Estimated annual AIDS related deaths','Estimated annual new HIV infections','Estimated number of people living with HIV','Estimated rate annual AIDS related deaths per 100000 people']]\
.groupby('Year', as_index=False)\
.diff().dropna()
diff_male_mean[['Country', 'Region', 'Age', 'Sex']] = ['Diff_male_mean', 'None', 'Age 10-19', 'Diff_male_mean']

diff_male_mean = diff_male_mean.reset_index(drop=True)
diff_male_mean['Year'] = mean_per_year['Year']
diff_male_mean.head(4)
```

### difference between female and male each year

#### difference-substract
outcoming dataframe "difference_female_male"

```python
print(mean_per_year_and_sex.columns)
mean_per_year_and_sex.head(4)
```

```python
difference_female_male = mean_per_year_and_sex\
.sort_values(by=['Year','Sex'])\
.groupby('Year',as_index=False)[['Estimated rate new HIV infection per 1000 people', 'Estimated annual AIDS related deaths','Estimated annual new HIV infections','Estimated number of people living with HIV', 'Estimated rate annual AIDS related deaths per 100000 people']].diff()

difference_female_male = difference_female_male.iloc[difference_female_male.index%2 == 1]
difference_female_male[['Country', 'Region', 'Age', 'Sex']] = ['Diff_female_male', 'None', 'Age 10-19', 'Diff_female_male']
difference_female_male = difference_female_male.reset_index(drop=True)
difference_female_male['Year'] = mean_per_year['Year']

difference_female_male.head(2)
```

#### difference-percentage
outcoming dataframes:
"percentage_of_female_to_male_per_year" and 
"percentage_of_male_to_female_per_year"

```python
mean_per_year_and_sex.head(4)
```

```python
#calculate percentage from female to male (female numbers/male numbers)
mean_per_year_and_sex = mean_per_year_and_sex.sort_values(by=['Year', 'Sex'])
percentage_of_female_to_male_per_year = get_percentage_from_row_above(mean_per_year_and_sex[['Estimated rate new HIV infection per 1000 people', 'Estimated annual AIDS related deaths','Estimated annual new HIV infections','Estimated number of people living with HIV','Estimated rate annual AIDS related deaths per 100000 people']])
percentage_of_female_to_male_per_year[['Country', 'Region', 'Age', 'Sex']] = ['Percentage_of_female_to_male', 'None', 'Age 10-19', 'Percentage_of_female_to_male']
percentage_of_female_to_male_per_year['Year'] = mean_per_year['Year']
percentage_of_female_to_male_per_year.head(4)
```

```python
#calculate percentage between male and female (male numbers/female numbers)
mean_per_year_and_sex = mean_per_year_and_sex.sort_values(by=['Year', 'Sex'])
percentage_of_male_to_female_per_year = get_percentage_from_row_below(mean_per_year_and_sex[['Estimated rate new HIV infection per 1000 people', 'Estimated annual AIDS related deaths','Estimated annual new HIV infections','Estimated number of people living with HIV','Estimated rate annual AIDS related deaths per 100000 people']])
percentage_of_male_to_female_per_year[['Country', 'Region', 'Age', 'Sex']] = ['Percentage_of_male_to_female', 'None', 'Age 10-19', 'Percentage_of_male_to_female']
percentage_of_male_to_female_per_year['Year'] = mean_per_year['Year']
percentage_of_male_to_female_per_year.head(4)
```

```python
def get_percentage_from_row_above(df):
    '''calculates percentage between the row above and the current 
    row for every second row- pecentage between female and male amount'''
    columns = df.columns
    temp_c = []
    df_percentage = pd.DataFrame()
    for c in columns:
        temp_l = []
        
        for l in range(0,len(df[c]),2):
            if l < len(df[c]):
                temp_l.append(round(df[c][l]/df[c][l+1], 3)*100)
        #temp_c.append(temp_l)
        df_percentage[c] = temp_l
    df_percentage.columns = df.columns
    return df_percentage
```

```python
def get_percentage_from_row_below(df):
    columns = df.columns
    temp_c = []
    df_percentage = pd.DataFrame()
    for c in columns:
        temp_l = []
        
        for l in range(0,len(df[c]),2):
            if l <= len(df[c]):
                temp_l.append(round(df[c][l+1]/df[c][l], 3))
        #temp_c.append(temp_l)
        df_percentage[c] = temp_l
    df_percentage.columns = df.columns
    return df_percentage
```

### concat all dataframes and save to csv

```python
data_and_total_numbers = pd.concat([data, mean_per_year_and_sex,mean_per_year,
                                    diff_female_mean, diff_male_mean, difference_female_male,
                                    percentage_of_male_to_female_per_year, percentage_of_female_to_male_per_year
                                   ])
data_and_total_numbers.to_csv('hiv_and_gender_inequality.csv')
data_and_total_numbers['Country'].value_counts()
```
